#### Multiple Linear Regression with (L1) Lasso Regularization Quiz

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
from sklearn.linear_model import Lasso
```

```python
# Read data into dataframe
train_data = pd.read_csv('/Users/elliotpartridge/Desktop/udacity_data_scientist_nanodegree/data_scientist_nanodegree_udacity/supervised_learning/linear_regression/linear_regression_quiz/data_lasso.csv', header=None)
train_data.head()
```

```python
x = train_data.iloc[:,:-1]
x.head()
```

```python
y = train_data.iloc[:,-1]
y.head()
```

```python
lasso_reg = Lasso()
lasso_reg.fit(x,y)
```

```python
reg_coef = lasso_reg.coef_
print(reg_coef)
```
